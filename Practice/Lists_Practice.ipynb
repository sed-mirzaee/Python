This is a sample code for practicing List operations.

We have a list of students.

```python
students = [
    {"name": "Ali", "score": 18},
    {"name": "Sara", "score": 12},
    {"name": "Reza", "score": 16},
    {"name": "Mina", "score": 20},
    {"name": "Ali", "score": 19}
]
```

# Print name of Students by FOR loop

```python
def print_name():
    for st in students:
        print(st["name"])

print_name()
```

# Create a new list and save just Scores by List comprehension

```python
def scores(condition=None):
    if condition is None:
        scores = [st["score"] for st in students]
    else:
        scores = [st["score"] for st in students if st["score"] > condition]

    return scores

#sc = scores()
sc = scores(16)
sc
```

# Check the pass status for each student

```python
def check_pass():
    for st in students:
        print(st["name"] + ":",  end=" ")
        if st["score"] >= 15 :
            print("Passed")
        else:
            print("Failed")

check_pass()
```


# Create a new List by List comprehesion for passed Students


```python
def passed_list(condition=None):
    if condition is None:
        passed_st = [st for st in students if st["score"] >= 15]
    else:
        passed_st = [st for st in students if st["score"] >= condition]

    return passed_st

ps = passed_list()
#ps = passed_list(16)
ps
```

# Create the **set** of student's name

We have douplicate here!

```python
def student_name():
    st_name = set()
    for st in students:
        st_name.add(st["name"])
    return st_name

stn = student_name()
stn
```

# Add new Student

```python
def student_add(name, score):
    students.append({"name": name, "score": score})

student_add("Amir", 14)
students
```

# Remove by name, remove all same name students.
! Note: For stop after first appearance, we need a break command.

```python
def student_remove(name):
    for st in students:
        if st["name"] == name:
            students.remove(st)

student_remove("Sara")
students
```

# Search by name


```python
def student_exist(name):
    for st in students:
        if st["name"] == name:
            return True

    return False

print("Is there a student with name 'Mina'? " + str(student_exist("Mina")))
print("Is there a student with name 'Sedigheh'? " + str(student_exist("Sedigheh")))
print("Is there a student with name 'Ali'? " + str(student_exist("Ali")))
```
